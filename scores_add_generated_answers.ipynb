Add generated answers from a file to the dataset.

```python
import json
def read_rows(path):
    rows = []
    for line in open(path):
        rows.append(json.loads(line))
    return rows

def write_json_format(path_out, rows):
    f_out = open(path_out, 'w')
    for row in rows:
        f_out.write(json.dumps(row, ensure_ascii=False)+'\n')
```

```python
path_dataset = 'test-B-big/dev.jl'
path_answers = 'answers.jl'
```

```python
rows = read_rows(path_dataset)
```

```python
answers = {}
for line in open(path_answers):
    pair = json.loads(line)
    answers[pair['question']] = pair['answer']
```

```python
for row in rows:
#     row['gpt3_answer'] = answers[row['question_translated']]
    row['chatgpt_answer'] = answers[row['question_text']]
```
